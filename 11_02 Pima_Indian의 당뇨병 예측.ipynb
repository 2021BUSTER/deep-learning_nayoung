```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import tensorflow as tf
```

```python
# seed 값 생성
# random 함수는 컴퓨터 안에 내장된 수많은 랜덤 테이블 중 하나를 불러내 그 표의 순서대로 숫자를 보여줌.
# seed 값을 설정하는 것은 랜덤 테이블 중 몇 번째 테이블을 불러올지 정하는 것
# -> seed 값이 같으면 같은 랜덤 값 출력!
# reproduction하기 위해 seed를 고정한다고 하는데 잘 이해가 안된다..

np.random.seed(3)
tf.random.set_seed(3)
```

```python
# 데이터 로드
dataset = np.loadtxt("../dataset/pima-indians-diabetes.csv",delimiter=",")
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
# 모델의 설정
model = Sequential()

# 3개의 층 생성
model.add(Dense(12,input_dim=8,activation='relu')) # 은닉층 + 입력층, 8개 입력데이터 가져옴, 12개 노드, 활성화 함수로 relu 사용
model.add(Dense(8,activation='relu')) # 은닉층, 8개 노드, 활성화 함수로 relu 사용
model.add(Dense(1,activation='sigmoid')) # 출력층, 1개 노드, 활성화 함수로 sigmoid 사용
```

```python
# 모델 컴파일
# 두 개의 예측값(당뇨병(1), 아님(0))을 갖기 때문에 오차함수로 binary_crossentropy 사용
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
# 모델 실행
model.fit(X,Y,epochs=200,batch_size=10)
```
